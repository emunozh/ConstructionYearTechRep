# 2. Neighbours ranking

```python
from scripts._getNeighbours import _getNeighbours
from scripts.plots import plotNeighbours
```

```python
from random import choice, seed
seed(12456)
ID = choice(Buildings.index)
```

```python
import numpy as np
```

```python
N = _getNeighbours(Buildings.loc[ID, "neighbours"], Buildings, ID)
```

```python
def fp(x):
    return "{:0.2f}".format(x)

def fi(x):
    return "{:}".format(x[-3:])

def fs(x):
    return "{:}".format(int(x))

N2 = N.ix[0:6]
N2.index = [fi(i) for i in N.ix[0:6].index]
    
with open("../TABLES/neighbours.tex", "w") as table:
    table.write(N2.loc[:,["baw", "gfk", "bja", "sqm",
                         "shell_wall", "distance", "rank"]].to_latex(formatters=[fs, fs, fs, fp, fp, fp , fp, fp]))
```

```python
plotNeighbours(Buildings, N, ID)
```

**Figure 3:** Ranking of neighbours based on building characteristics and euclidean distance
